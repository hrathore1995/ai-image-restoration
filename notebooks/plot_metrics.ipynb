```python
import pandas as pd
import matplotlib.pyplot as plt

# Load metrics
metrics_file = "../results/metrics_all.csv"
df = pd.read_csv(metrics_file)

# Drop rows with missing values
df = df.dropna()

# Compute averages
methods = ["Telea", "Navier", "Stable Diffusion"]

mean_psnr = [
    df["telea_psnr"].mean(),
    df["navier_psnr"].mean(),
    df["sdxl_psnr"].mean()
]

mean_ssim = [
    df["telea_ssim"].mean(),
    df["navier_ssim"].mean(),
    df["sdxl_ssim"].mean()
]

mean_lpips = [
    df["telea_lpips"].mean(),
    df["navier_lpips"].mean(),
    df["sdxl_lpips"].mean()
]

# --- Plot PSNR ---
plt.figure(figsize=(7,5))
plt.bar(methods, mean_psnr, color=["#1f77b4", "#ff7f0e", "#2ca02c"])
plt.title("Average PSNR by Method")
plt.ylabel("PSNR (dB)")
plt.ylim(0, max(mean_psnr) + 5)
for i, v in enumerate(mean_psnr):
    plt.text(i, v + 0.3, f"{v:.2f}", ha="center", fontsize=10)
plt.show()

# --- Plot SSIM ---
plt.figure(figsize=(7,5))
plt.bar(methods, mean_ssim, color=["#1f77b4", "#ff7f0e", "#2ca02c"])
plt.title("Average SSIM by Method")
plt.ylabel("SSIM (01)")
plt.ylim(0, 1)
for i, v in enumerate(mean_ssim):
    plt.text(i, v + 0.02, f"{v:.3f}", ha="center", fontsize=10)
plt.show()

# --- Plot LPIPS ---
plt.figure(figsize=(7,5))
plt.bar(methods, mean_lpips, color=["#1f77b4", "#ff7f0e", "#2ca02c"])
plt.title("Average LPIPS by Method")
plt.ylabel("LPIPS (lower = better)")
plt.ylim(0, max(mean_lpips) + 0.1)
for i, v in enumerate(mean_lpips):
    plt.text(i, v + 0.01, f"{v:.3f}", ha="center", fontsize=10)
plt.show()
```
